```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import text
from nltk.corpus import stopwords
import nltk

# Paso 1: Lista de textos (puedes reemplazar con reseñas reales)
resenas = [
    "El lugar es abierto  y pues muy bien ubicado. El lugar como tal bien , solo un poco incómodas las mesas ... y pues un poco descuidado  el inmobiliario.  El servicio  por parte de una extranjera ( tal vez cubana) muy amable y pues la comida  muy muy buena. Los tacos realmente  buenos . Muy recomendable...",
    "Los tacos con queso en extremo grasosos. El pastor bien, sin pena ni gloria. El pan árabe NO lo recomiendo para nada, es caro para la porción que sirven. Las cebollitas bien, están ricas. Los frijoles carros están buenos. Las costras de queso aunque grasosas, están bien en porción-precio-sabor. Lástima de salsas no hay una que pique. Todas están muy flojas y no aportan nada al taco. El alambre está bueno. El servicio medio medio, un poco despistados y los refrescos al tiempo cuando se pidieron fríos.",
    "Un poco lento el servicio,  aunqur cabe mencionar que tenían mucha gente. Ame el queso, muy delicioso. Porciones pequeñas,  pero rico, insisto el queso sumamente delicioso y recomendado.",
    "Taqueria ubicada en la volonia valle norte en la calle de Parroquia. El menú es amplio. Destacan los volcanes y los alambres. Sin faltar los imperdibles tacos de pastos. Tambien hay sopas y entradas, no pudiendo faltar los quesos fundidos. Servicio atento y rápido.  Precios accesibles. Hay aguas frescas y cervezas.",
    "Asquerosa comida! Pedí dos tacos, uno de bistec y otro de arrachera, los pedí bien cocidos y me los trajeron crudos, las tortillas se rompen, no están bien calentadas y no te sirven con absolutamente nada el taco, no tiene cebolla, cilantro, nopales, papas o algún acompañamiento. No son baratos y no me quejo, pero pagar 65 pesos por un mal taco duele. Nada recomendable y el servicio también es deplorable, los meseros son mal modientos. Pedí un guacamole que ni sal tenía, es mas solo encontré 1 hojita de cilantro",
    "Hay diversos puntos para tomar mesa: dentro del lugar, a pie de banqueta o sobre la calle que está techado.La carta tiene suficientes platillos para escoger entre alambres, gringas, tacos, costras o volcanes; su sazón ha mejorado al igual que su servicio, ya que les di una segunda oportunidad porque la primera vez había dejado mucho a desear.El sanitario es uno solo pero en general procuran mantener limpio y ordenado, los precios son accesibles para la zona en la que se encuentra el lugar.Definitivamente recomiendo que hagan la prueba al menos una vez, no se arrepentirán, 😋🤙🏼"
]

# Paso 1.1: Asegúrate de tener las stopwords descargadas
nltk.download('stopwords')
stopwords_es = stopwords.words('spanish')

# Paso 1.2: Agregamos nuestras propias palabras personalizadas
palabras_adicionales = ['pues', 'bien', 'vez', 'tal', 'pedí', 'taco', 'tacos', 'lugar']  # ← aquí tus palabras específicas
stopwords_es.extend(palabras_adicionales)

# Paso 2: Vectorizar con TF-IDF
vectorizador = TfidfVectorizer(stop_words=stopwords_es)
X = vectorizador.fit_transform(resenas)

# Paso 3: Obtener los términos con mayor puntuación promedio
promedios = X.mean(axis=0).A1
palabras = vectorizador.get_feature_names_out()

# Creamos el DataFrame
resultados = pd.DataFrame({'palabra': palabras,'tfidf': promedios})
# Ordenamos por puntuación tfidf descendente y mostramos las 10 más relevantes
resultados = resultados.sort_values(by='tfidf', ascending=False).head(10).reset_index(drop=True)

# Paso 4: Mostrar los 10 temas más representativos
print("🔍 Palabras clave más representativas:")
print(resultados.head(10))
```
